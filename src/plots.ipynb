```python
import matplotlib.pyplot as plt
plt.rc('legend',fontsize=16)
labels_fontsize=16
lwd = 3
ms = 8

factors = ["8", "16", "32", "64"]

neumf_org_hr = [.688, .696, .701, .705]
neumf_org_ndcg = [.410, .420, .425, .426]

neumf_hr = [.671, .700, .707, .710]
neumf_ndcg = [.400, .425, .432, .444]

neumf_attn_hr = [.686, .725, .721, .704] # ostatnia do uzupełnienia
neumf_attn_ndcg = [.404, .452, .454, .434] # ostatnia do uzupełnienia
```

```python
import numpy as np
hr = np.array(neumf_attn_hr)
hr_org = np.array(neumf_hr)
ndcg = np.array(neumf_attn_ndcg)
ndcg_org = np.array(neumf_ndcg)

hr_changes = (hr-hr_org)/hr_org*100
ndcg_changes = (ndcg-ndcg_org)/ndcg_org*100

hr_changes = [round(h, 2) for h in hr_changes]
ndcg_changes = [round(h, 2) for h in ndcg_changes]

print(hr_changes)
print(ndcg_changes)
```
